# Are the Healthy Breakfast Cereals Healthy?

Data set source: Kaggle data 
- gathered and cleaned up by Petra Isenberg, Pierre Dragicevic and Yvonne Jansen for Kaggle
- originally from Carnegie Mellon University stats libarary - 1993 Statistical Graphics Exposition

Dataset descriptioN: This is a multivariate dataset describing seventy-seven
commonly available breakfast cereals, based on the information now available on
the newly-mandated F&DA food label.  What are you getting when you eat a bowl
of cereal?  Can you get a lot of fiber without a lot of calories?  Can you
describe what cereals are displayed on high, low, and middle shelves?  The good
news is that none of the cereals for which we collected data had any
cholesterol, and manufacturers rarely use artificial sweeteners and colors,
nowadays.  However, there is still a lot of data for the consumer to understand
while choosing a good breakfast cereal.

# Examine nutritional profiles with PCA

```python
cols_for_pca
```

```python
# Initialize a new PCA model with a default number of components.
pca = sklearn.decomposition.PCA()
variances = []
krange = list(range(1,(len(cols_for_pca)+1),1))
print("Different # of components to try:", krange)
print(len(krange))
```

```python
# calculate the explained variance for different # of components
for k in krange:
    pca_object = PCA(n_components=k, random_state=1)
    pca_object.fit(X_cols)
    variances.append(pca_object.explained_variance_ratio_.sum()) # cumulative variance explained
```

```python
# Plot k's vs explained variance
plt.plot(krange, variances[:], '-o')
plt.xlabel('number of components, k')
plt.ylabel('cumulative explained variance')
plt.title('Explained Variance vs Number of Components')
plt.xticks(range(1,max(krange),1))
plt.show()
```

```python
pca_5comp_object = PCA(n_components=5, random_state=1)
X_pca = pca_5comp_object.fit_transform(X_cols)
print("The % variance the components explain:", pca_5comp_object.explained_variance_ratio_)
```

```python
print(f"The 2 components explain: {round(sum(pca_5comp_object.explained_variance_ratio_[:4]),2)} of the variation in data")
```

```python
print("Component attribute shows how each component shows how each of the features contributes to each component:")
print(pca_5comp_object.components_.shape)
```

```python
print(f"Shape of transformed data: {X_pca.shape}")  # Should be (n_samples, 2)
```

# Kmeans clustering algorithm to see intrinsic groups

```python
ssws = []
krange = range(1,12)

for k in krange:
    model = KMeans(n_clusters =k, n_init='auto', init='random')
    model.fit(X_cols)
    ssws.append(model.inertia_)
```

```python
# Plot k's vs inertias, show just last 10 values of k range
plt.plot(krange[-10:], ssws[-10:], '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(krange)
plt.show()
```

Data seems to have an intrinsic dimension of 6 clusters

```python
# re-run with 6 clusters
KMeans_model = KMeans(n_clusters=6, random_state=10, n_init='auto', init='random')
cluster6_labels = KMeans_model.fit_predict(X_cols)
```

# Notes

**questions/problems to communicate:** 
- are all the cereals the same? are they all healthy?
- a few of them are unusual in some aspects
- healthy names don't mean much

**Key messages/solutions**


* Can you find the correlations you might expect?  Are there any surprising
correlations? 
* What is the true "dimensionality" of the data?
* Are there any cereals which are virtually identical?
* Is there any way to discriminate among the major manufacturers by cereal
characteristics, or do they each have a "balanced portfolio" of cereals?
* Do the nutritional claims made in cereal advertisements stand the scrutiny of
data analysis? 
* Are there cereals which are clearly nutritionally superior, or inferior?  Are
there clusters of cereals? 
* Is a ranking or scoring scheme possible or reasonable, and if so, are there
cereals which are nutritionally superior or inferior under all reasonable
weighting schemes?
